```python
from scipy.stats import ttest_1samp
import numpy as np
ages = [23, 34, 65, 12, 34, 56, 98, 67, 28, 48]
print(ages)
mean = np.mean(ages)
print(mean)
stddev = np.std(ages)
t_test, p_val = ttest_1samp(ages, 40)
print("t-test value", t_test)
print("p-value", p_val)
if p_val<0.05:
  print("rejecting null hypothesis")
else:
 print("accepting null hypothesis")
```

```python
from scipy.stats import ttest_ind
data_group1 = np.array([12, 18, 12, 13, 15, 1, 7,
                        20, 21, 25, 19, 31, 21, 17,
                        17, 15, 19, 15, 12, 15])
data_group2 = np.array([23, 22, 24, 25, 21, 26, 21,
                        21, 25, 30, 24, 21, 23, 19,
                        14, 18, 14, 12, 19, 15])
t_test, p_val = ttest_ind(data_group1, data_group2)
print("t-test", t_test)
print("p_val", p_val)
if p_val<0.05:
  print("rejecting null hypothesis")
else:
  print("accepting null hypothesis")
```

```python
import pandas as pd
df = pd.read_csv("netflix_titles.csv")
no_director = df[df['director'].isnull()].index
df2 = df.drop(no_director, axis = 0)
```

```python
df2['duration'].fillna(0, inplace = True)
df2.isnull().sum()
df_movie = df2[df2['type']=='Movie']
df_movie = df_movie.assign(minute = df_movie['duration'].str.extract(r'(\d+)', expand = False).astype(float))
```

```python
import math
min_minute, max_minute = df_movie['minute'].min(), df_movie['minute'].max()
minbin, maxbin = math.floor(min_minute), math.ceil(max_minute)
gap = maxbin - minbin
bins = 10
width = math.ceil(gap/bins)
bins = np.arange(minbin, maxbin+width, width)
fig, ax = plt.subplots(nrows = 1, ncols = 1)
ax.hist(df_movie['minute'], bins = bins, color = 'turquoise')
fig.set_size_inches(10, 6)
ax.set_xlabel('Duration (minutes)', fontsize=15)
ax.set_ylabel('Number of Movies', fontsize=15)
plt.xticks(bins, fontsize = 15)
plt.yticks(fontsize = 15)
fig.tight_layout()
plt.show()
```
