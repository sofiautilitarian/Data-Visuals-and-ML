```python
import pandas as pd
df = pd.read_csv("marketing_campaign.csv")
df.head()
```

```python
r1 = df[df['Response']==1]
```

```python
r2 = df[df['Response']==0]
```

```python
rec1 = r1['Recency']
rec2 = r2['Recency']
```

```python
r1_sampled = rec1.sample(n=20, random_state=120)
r2_sampled = rec2.sample(n=20, random_state=120)
```

```python
r1list = r1_sampled.tolist()
r2list = r2_sampled.tolist()
```

```python
from scipy.stats import ttest_ind
# Implementing the t-test
t_test,p_val = ttest_ind(r1list, r2list)
print("The t-test is: ", t_test)
print("The P-value is: ", p_val)

# taking the threshold value as 0.05 or 5%
if p_val < 0.05:
    print("Relation exists between response and recency")
else:
    print("no relation between response and recency")
```

```python
data1 = df[df['Kidhome']==0]
data2 = df[df['Kidhome']==1]
data3 = df[df['Kidhome']==2]
```

```python
import scipy.stats
new1 = data1['NumWebPurchases']
new2 = data2['NumWebPurchases']
new3 = data3['NumWebPurchases']

new1_sampled = new1.sample(n=20, random_state=120).tolist()
new2_sampled = new2.sample(n=20, random_state=120).tolist()
new3_sampled = new3.sample(n=20, random_state=120).tolist()
# Performing the F-Test
f_test, p_val = scipy.stats.f_oneway(new1_sampled, new2_sampled, new3_sampled)

print("f-test value is: ", f_test)
print("p-value is: ", p_val)

# taking the threshold value as 0.05 or 5%
if p_val < 0.05:
    print(" We can reject the null hypothesis")
else:
    print("We can accept the null hypothesis")
```
